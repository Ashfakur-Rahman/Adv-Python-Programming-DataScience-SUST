# Summary Statistics

```python
import numpy as np
import pandas as pd
from scipy import stats

# Sample data
data = [12, 15, 14, 10, 18, 20, 20, 22, 15, 16]

# Mean
mean_value = np.mean(data)
print("Mean:", mean_value)

# Median
median_value = np.median(data)
print("Median:", median_value)

# Mode
mode_value = stats.mode(data, keepdims=True)
print("Mode:", mode_value.mode[0])

# Range
data_range = np.max(data) - np.min(data)
print("Range:", data_range)

# Variance
variance_value = np.var(data, ddof=1)   # sample variance
print("Variance:", variance_value)

# Standard Deviation
std_dev = np.std(data, ddof=1)
print("Standard Deviation:", std_dev)

# Interquartile Range (IQR)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
print("IQR:", iqr)

# Skewness
skewness = stats.skew(data)
print("Skewness:", skewness)

# Kurtosis
kurtosis = stats.kurtosis(data)
print("Kurtosis:", kurtosis)
```

# Visualizing Summary Statistics

```python
import matplotlib.pyplot as plt
import seaborn as sns

# For consistent style
sns.set(style="whitegrid")

# Histogram
plt.figure(figsize=(6,4))
sns.histplot(data, kde=False, bins=8)
plt.title("Histogram")
plt.show()

# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x=data)
plt.title("Boxplot")
plt.show()

# Bar Chart (categorical example)
categories = ['A', 'B', 'C']
counts = [10, 25, 15]

plt.figure(figsize=(6,4))
sns.barplot(x=categories, y=counts)
plt.title("Bar Chart")
plt.show()

# Density Plot
plt.figure(figsize=(6,4))
sns.kdeplot(data, shade=True)
plt.title("Density Plot")
plt.show()
```

**Advanced Visualization Techniques**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 1: Generate random numeric data ---
np.random.seed(42)
df = pd.DataFrame(np.random.randn(150, 4), columns=["Var A", "Var B", "Var C", "Var D"])

# --- Optional: Add a categorical variable for FacetGrid / hue ---
df['Category'] = np.random.choice(['Group 1', 'Group 2'], size=len(df))

# --- Step 2: Histogram (distribution) ---
plt.figure(figsize=(6,4))
sns.histplot(df["Var A"], bins=15, kde=True, color="skyblue")
plt.title("Histogram of Var A")
plt.tight_layout()
plt.show()

# --- Step 3: Boxplot (spread + outliers) ---
plt.figure(figsize=(6,4))
sns.boxplot(data=df[["Var A","Var B","Var C","Var D"]])  # removed palette to avoid warning
plt.title("Boxplot of All Variables")
plt.tight_layout()
plt.show()

# --- Step 4: Heatmap (correlation) ---
plt.figure(figsize=(6,4))
sns.heatmap(df[["Var A","Var B","Var C","Var D"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

# --- Step 5: Pairplot (scatterplot matrix with hue) ---
sns.pairplot(df[["Var A","Var B","Var C","Var D","Category"]], hue="Category", diag_kind="kde", palette="bright")
plt.suptitle("Pairplot with Hue", y=1.02)
plt.show()

# --- Step 6: Violin Plot (distribution + density) ---
plt.figure(figsize=(6,4))
sns.violinplot(data=df[["Var A","Var B","Var C","Var D"]])  # removed palette to avoid warning
plt.title("Violin Plot of Variables")
plt.tight_layout()
plt.show()

# --- Step 7: Facet Grid Example ---
g = sns.FacetGrid(df, col="Category")  # works because 'Category' exists
g.map_dataframe(sns.histplot, x="Var A", bins=10, kde=True, color="orange")
g.set_titles("Category: {col_name}")
plt.tight_layout()
plt.show()
```
