**এই নোটবুকে আমরা মূলত Python দিয়ে কিভাবে Statistical summary এবং তাদের Visualization করতে হয় তা Practice করবো!**

```python
# এখানে প্রয়োজনীয় লাইব্রেরি গুলো ইমপোর্ট (লোড) করা হচ্ছে

import numpy as np          # সংখ্যাগত হিসাব, array ও matrix অপারেশনের জন্য NumPy
from scipy import stats     # পরিসংখ্যান সংক্রান্ত কাজ (mean, median, hypothesis test ইত্যাদি) করার জন্য stats মডিউল
```

```python
# এখন আমরা একটি sample ডেটা নিচ্ছি
# এটি পাইথনের একটি list, যেখানে কিছু সংখ্যাগত মান (numerical values) রাখা হয়েছে
# এই ডেটা ব্যবহার করে আমরা পরবর্তীতে বিভিন্ন পরিসংখ্যানগত হিসাব করবো
data = [12, 15, 14, 10, 18, 20, 20, 22, 15, 16]
#================================================================================================================

# এখন আমরা উপরের list data এর Mean (গড় মান) হিসাব করবো
mean_value = np.mean(data)    # np হলো NumPy লাইব্রেরির সংক্ষিপ্ত নাম
                              # mean() ফাংশনটি list এর সব মান যোগ করে মোট সংখ্যার দ্বারা ভাগ করে গড় বের করে
                              # mean_value একটি variable, যেখানে Mean সংরক্ষণ করা হয়েছে
# এবার আমরা Mean এর মানটি স্ক্রিনে দেখাবো
print("Mean: ",mean_value)             # print() ফাংশন ব্যবহার করে Mean এর মান ডিসপ্লেতে দেখানো হচ্ছে
#================================================================================================================

# এখন আমরা উপরের list data এর Median (মধ্যক) হিসাব করবো
median_value = np.median(data)   # np.median() ডেটাগুলো sort করে মাঝের মান বের করে
                                 # যদি মোট ডেটা সংখ্যা জোড় হয়, তাহলে মাঝের দুইটির গড় নেয়

# Median এর মান স্ক্রিনে দেখানো হচ্ছে
print("Median: ",median_value)              # print() দিয়ে Median ডিসপ্লেতে দেখানো
# ===============================================================================================================

# এখন আমরা উপরের list data এর Mode (সবচেয়ে বেশি frequent মান) হিসাব করবো
# stats.mode() ফাংশন ব্যবহার করা হচ্ছে
# এটি Scipy লাইব্রেরির stats মডিউল থেকে নেওয়া হয়েছে
# ডেটার মধ্যে সবচেয়ে বেশি বার প্রাপ্ত মান (most frequent value) বের করে
# keepdims=True ব্যবহার করলে output array আকারে আসে এবং warning আসে না
mode_value = stats.mode(data, keepdims=True)

# Mode এর মান স্ক্রিনে দেখানো হচ্ছে
# mode_value.mode → array আকারে Mode রাখে
# [0] ব্যবহার করে array থেকে প্রথম মান (যা সবচেয়ে frequent) আমরা বের করছি
print("Mode:", mode_value.mode[0])

# Mode কতবার এসেছে তা দেখানো হচ্ছে
# mode_value.count → array আকারে frequency রাখে
# [0] ব্যবহার করে প্রথম frequency বের করছি
print("Frequency of Mode:", mode_value.count[0])
#==============================================================================================

# চলুন এবার আমরা Range হিসাব করবো
# Range মানে হলো ডেটার সর্বোচ্চ মান এবং সর্বনিম্ন মানের পার্থক্য
# np.max(data) → ডেটার সর্বোচ্চ মান বের করে
# np.min(data) → ডেটার সর্বনিম্ন মান বের করে
# তাদের পার্থক্য নিলে Range পাওয়া যায়
data_range = np.max(data) - np.min(data)

# Range এর মান স্ক্রিনে দেখানো হচ্ছে
print("Range:", data_range)
#====================================================================================================

# এখন আমরা Variance (বিচ্যুতি / variance) হিসাব করবো
# Variance হলো ডেটার মানগুলি কতটা গড় মান থেকে বিচ্যুত হয়েছে তা পরিমাপ করার একটি পদ্ধতি
# np.var() ব্যবহার করে Variance বের করা হয়
# ddof=0 → population variance (default)
# ddof=1 → sample variance
variance_value = np.var(data, ddof=0)   # population variance

# Variance স্ক্রিনে দেখানো হচ্ছে
print("Variance: ",variance_value)
#====================================================================================================

# Now Standard Deviation (SD / মানদণ্ড বিচ্যুতি) হিসাব
# Standard Deviation হলো Variance এর square root
# এটি ডেটার মান গড় থেকে কতটা ছড়ানো আছে তা নির্দেশ করে
# np.std() ব্যবহার করে Standard Deviation বের করা হয়
# ddof=0 → population SD (default)
# ddof=1 → sample SD
std_value = np.std(data, ddof=0)   # population standard deviation

# Standard Deviation স্ক্রিনে দেখানো হচ্ছে
print("Standard Deviation: ",std_value)
# ===========================================================================================================

# Now Interquartile Range (IQR) হিসাব
# IQR হলো ডেটার Q3 (তৃতীয় quartile) এবং Q1 (প্রথম quartile) এর মধ্যে পার্থক্য
# Q1 → ডেটার প্রথম 25% মানের মধ্যবর্তী মান
# Q3 → ডেটার প্রথম 75% মানের মধ্যবর্তী মান
# IQR = Q3 - Q1
Q1 = np.percentile(data, 25)   # 25th percentile → প্রথম quartile
Q3 = np.percentile(data, 75)   # 75th percentile → তৃতীয় quartile
iqr_value = Q3 - Q1             # Interquartile Range

# IQR স্ক্রিনে দেখানো হচ্ছে
print("IQR: ",iqr_value)
# ===============================================================================================

# Skewness হিসাব
# Skewness হলো ডেটার distribution এর symmetry measure
# যদি skewness = 0 → perfectly symmetric distribution
# যদি skewness > 0 → right-skewed (long tail ডান দিকে)
# যদি skewness < 0 → left-skewed (long tail বাম দিকে)
# Scipy stats এর skew() ফাংশন ব্যবহার করা হয়
skewness_value = stats.skew(data)

# Skewness স্ক্রিনে দেখানো হচ্ছে
print("Skewness: ",skewness_value)
# ====================================================================================================

# Now Kurtosis হিসাব
# Kurtosis হলো distribution এর "peakedness" বা শীর্ষতার measure
# যদি kurtosis ≈ 0 → Normal distribution এর মত shape (mesokurtic)
# যদি kurtosis > 0 → sharper peak, heavy tails (leptokurtic)
# যদি kurtosis < 0 → flatter peak, light tails (platykurtic)
# Scipy stats এর kurtosis() ফাংশন ব্যবহার করা হয়
kurtosis_value = stats.kurtosis(data)

# Kurtosis স্ক্রিনে দেখানো হচ্ছে
print("Kurtosis: ",kurtosis_value)
```

**এবার আমরা Visualize করার চেষ্টা করবো!!**

```python
# Visualization এর জন্য প্রয়োজনীয় লাইব্রেরি ইমপোর্ট
# গ্রাফ, চার্ট ও প্লট আঁকার জন্য matplotlib লাইব্রেরি ইমপোর্ট করা হচ্ছে
import matplotlib.pyplot as plt

# ডেটার উপর ভিত্তি করে সুন্দর ও স্ট্যাটিস্টিক্যাল গ্রাফ দেখানোর জন্য seaborn লাইব্রেরি ইমপোর্ট করা হচ্ছে
import seaborn as sbn

# গ্রাফের স্টাইল একইরকম ও সুন্দর রাখার জন্য
# সব গ্রাফে একরকম পরিষ্কার ও গ্রিড-ভিত্তিক স্টাইল ব্যবহার করার জন্য seaborn এর স্টাইল সেট করা হচ্ছে
sbn.set(style="darkgrid") #সুজা বাংলায় গ্রাফগুলোর ব্যাকগ্রাউন্ড সাদা এবং গ্রিড লাইন থাকবে!
```

**1. Histogram**

```python
# গ্রাফের সাইজ সেট করা (প্রস্থ=6 ইঞ্চ, উচ্চতা=4 ইঞ্চ)
plt.figure(figsize=(6,4))

# Histogram আঁকা
# data → যেই ডেটা দিয়ে histogram বানানো হবে
# bins=8 → ডেটাকে ৮টি interval/bar-এ ভাগ করা হবে, এটার value চেঞ্জ করে বার গুলো গুনে দেখতে পারেন!
# kde=False → smooth curve বন্ধ রাখা হয়েছে, শুধু বার দেখানো হবে, এটার value True করে দিয়ে চেঞ্জ দেখলেই বুঝে যাবেন!
sbn.histplot(data, kde=False, bins=8)

# গ্রাফের শিরোনাম (title) দেওয়া হচ্ছে
plt.title("Histogram")

# চাইলে X label এবং Y label ও সেট করতে পারবেন
plt.xlabel("Value")
plt.ylabel("Frequency")

# গ্রাফ স্ক্রিনে দেখানো হচ্ছে, এটা প্রত্যেক গ্রাফে মাস্ট
plt.show()

```

**2. Box Plot**

```python
plt.figure(figsize=(6,4)) #see previous graph
sbn.boxplot(x=data)       # বাই ডিফল্ট এটা y axis বরাবর থাকে, x=data লিখে horizontally ঘুরিয়ে নিলাম
plt.title("Boxplot")      #see previous graph
plt.show()                #see previous graph
```

**3. Bar Chart(categorical Example)**

```python
# ধরো আমরা দুইটা ডাটা নিলাম, একটা students দের নাম আর আরেকটা তাদের Marks
students = ['A', 'B', 'C']
marks = [10, 25, 15]

plt.figure(figsize=(6,4))
sbn.barplot(x=students, y=marks) # x axis এর দিকে students এবং y axis এর দিকে marks রাখলাম
plt.title("Bar Chart")
plt.show()
```

**4. Density Plot**

```python
plt.figure(figsize=(6,4))
sbn.kdeplot(data, fill=True) # fill = False করলেই বুঝতে পারবেন বিষয়টা
plt.title("Density Plot")
plt.show()
```
