# SOLUTION

## Question 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import (
    LinearRegression,
    LogisticRegression
 )
print('Successfully imported!!')
```

```python
# 1
data = {
    'Student_ID': [1, 2, 3, 4, 5],
    'Hours_Studied': [2, 3, 5, 7, 9],
    'Exam_Score': [45, 50, 65, 78, 92],
    'Passed': [0, 0, 1, 1, 1]
}
df = pd.DataFrame(data)
print('DataFrame is Created!!')
print(df)
```

```python
# 2
mean_hours = df['Hours_Studied'].mean()
std_hours = df['Hours_Studied'].std()
print(f'Mean of Hours_Studied: {mean_hours}')
print(f'Standard Deviation of Hours_Studied: {std_hours}')
```

```python
# 3
plt.scatter(df['Hours_Studied'], df['Exam_Score'])
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_Score')
plt.title('Scatter Plot')
plt.show()
```

## Question 2

```python
# 1
X = df[['Hours_Studied']].values        # features variable
y = df['Exam_Score'].values             # Target variable

model = LinearRegression()
model.fit(X, y)
print('Model is Trained!!')
```

```python
# 2
hours_studied = np.array([[6]])
predicted_score = model.predict(hours_studied)
print(f'Predicted Score for {hours_studied} hours studied: {predicted_score[0]:.2f}')
```

## Question 3

```python
# 1
X = df[['Hours_Studied']].values        # features variable
y = df['Passed'].values                 # Target variable

model = LogisticRegression()
model.fit(X, y)
print('Model is Trained!!')
```

```python
# 2
hours_studied = np.array([[4]])
predicted_outcome = model.predict(hours_studied)
print(f'Predicted Outcome for {hours_studied} hours studied: {predicted_outcome[0]}')

if(predicted_outcome[0] == 1):
    print('Student will Pass')
else:
    print('Student will Fail')
```
