# Linear Regression (Problem Definition)
ধরো তোমার একটা আইসক্রিমের দোকান আছে এবং তুমি লক্ষ্য করলে যে- প্রতিদিনের তাপমাত্রার উপর তোমার আইসক্রিম বিক্রি কিছুটা সম্পর্কিত। সাধারণত তাপমাত্রা বাড়লে আইসক্রিম বেশি বিক্রি হয় আর তাপমাত্রা কমলে আইসক্রিম কম বিক্রি হয়। তাহলে তোমার কাছে একটা ডাটাসেট আছে যেখানে প্রতিদিনের temperature এর সাথে ঐদিনের sales এর পরিমাণ উল্লেখ করা আছে।!dataset.jpg

## Loading Libraries/Modules

```python
# Numerical Python লাইব্রেরি ইমপোর্ট করা হচ্ছে, যা array এবং গণনার জন্য ব্যবহৃত হয়
import numpy as np

# scikit-learn লাইব্রেরি থেকে LinearRegression ক্লাস ইমপোর্ট করা হচ্ছে
# এটি আমাদেরকে লিনিয়ার রিগ্রেশন মডেল তৈরি করতে সাহায্য করবে
from sklearn.linear_model import LinearRegression
```

## Know the DATASET!

```python
# X হলো Input (Temperature), y হলো Output (Sales)

# ইনপুট ডেটা: temperature (℃)
# এখানে ডাবল ব্র্যাকেট [[ ]] ব্যবহার করা হয়েছে কারণ scikit-learn কলাম আকারে ডেটা চায়
X = np.array([[20], [25], [30], [35], [40]])

# আউটপুট ডেটা: sales ($)
# সব মানকে list আকারে দিতে হবে
y = np.array([200, 300, 400, 500, 600])
```

## Model Selection and Training

```python
model = LinearRegression() # model is created
model.fit(X,y)             # fit/train the model on our data
print("Model is Trained!") # After training just show the text
```

## PREDICT FOR A NEW TEMPERATURE!!

```python
# Predict for 32 degrees
predicted_result = model.predict([[32]])      # predict the unknown sales for the temperature 32 degrees through your model
print("Predicted Result: ", predicted_result) # display the sales for 32 degrees
```
